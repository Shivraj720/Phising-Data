## Splitting data into train & test part

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### As the problem is of classification & output varibale has 2 classes, must have to use classification algorithms

```python
from sklearn.linear_model import LogisticRegression
```

```python
l=LogisticRegression()
```

```python
l.fit(x_train,y_train)
```

```python
y_pred=l.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print("Accuracy of Logistic is:",accuracy_score(y_test,y_pred)*100)
```

```python
print("Confusion_matrix for Logistic is:\n",confusion_matrix(y_test,y_pred))
```

```python
print("Classification_report of Logistic is:\n",classification_report(y_test,y_pred))
```
